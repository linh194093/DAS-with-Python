```python
def mergeSort(list, start, end):
    if start < end:
        arg = start + (end-start)//2

        mergeSort(list, start, arg)
        mergeSort(list, arg + 1, end)
        merge(list, start, arg, end)
 
```

```python
def merge(list, start, arg,  end):
    n1 = arg - start + 1
    n2 = end - arg
    #Create 2 temp
    list1 = [0] * (n1)
    for i in range(0, n1):
        list1[i] = list[start + i]

    list2 = [0] * (n2)
    for j in range(0, n2):
        list2[j] = list[arg + 1 + j]

    i = 0
    j = 0
    k = start

    while i < n1 and j < n2:
        if list1[i] <= list2[j]:
            list[k] = list1[i]
            i += 1
        else:
            list[k] = list2[j]
            j += 1
        k += 1

    #scrape
    while i < n1:
        list[k] = list1[i]
        i += 1
        k += 1
    while j < n2:
        list[k] = list2[j]
        j += 1
        k += 11
```
